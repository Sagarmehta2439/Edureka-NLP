Count the number of articles in file ‘Macbeth.txt’ (Articles are ‘a’, ‘an’ &’the’ ) and save the file as ‘Macbeth-ArticlesRemoved.txt’

```python
with open('Macbeth.txt','r') as file:
    mb = file.read()
```

```python
len(mb)
```

```python
mb_tokenize = nltk.word_tokenize(mb)
```

```python
mb_tokenize
```

```python
from nltk.probability import FreqDist
import re
```

```python
fdist = FreqDist()
```

```python
for word in mb_tokenize:
    fdist[word.lower()] += 1
fdist
```

```python
Articles = ['a','an','the']
for article in mb_tokenize:
    if article in Articles:
        mb = article.drop()
        print(mb)
```

10.Remove all the punctuations from the file. (Punctuations are . ,’ : ; # ( ) etc.) and save as ‘Macbeth-punctuationsRemoved.txt’

```python
import re
punctuation = re.compile(r'[-.?!,:;()|0-9]')
post_punctuation = []
for words in mb_tokenize:
    word = punctuation.sub("",words)
    if len(word)>0:
        post_punctuation.append(word)
```

```python
len(post_punctuation)
```
